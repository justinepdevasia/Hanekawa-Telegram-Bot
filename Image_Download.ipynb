```
import requests
import time
import concurrent.futures


fp=open("no_download.txt","w")

def download_image(img_url,name):
    img_resp = requests.get(img_url)
    if img_resp.status_code==200:
      img_bytes=img_resp.content
      img_name = name
      if img_url[-3:]=="jpg":
        img_name = f'data/{img_name}.jpg'
        with open(img_name, 'wb') as img_file:
          img_file.write(img_bytes)
          print(f'{img_name} was downloaded...')
      elif img_url[-3:]=="png":
        img_name = f'data/{img_name}.png'
        with open(img_name, 'wb') as img_file:
          img_file.write(img_bytes)
          print(f'{img_name} was downloaded...')
      else:
        print(f"{name} not downloadable")
        fp.write(str(name)+"\n")
    else:
      print(f"{name} not downloadable")
      fp.write(str(name)+"\n")


t1 = time.perf_counter()

with concurrent.futures.ThreadPoolExecutor() as executor:
    for name,url in enumerate(urls):
        executor.submit(download_image, url, name+1)

t2 = time.perf_counter()

print(f'Finished in {t2-t1} seconds')
fp.close()
```
